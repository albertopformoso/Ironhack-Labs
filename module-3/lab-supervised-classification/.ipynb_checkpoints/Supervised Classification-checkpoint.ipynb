La columna 'e' nos indica el char al que hace referencia la fila por lo que del 0 al 9 tenemos números (0-9), del 10 al 35 las mayúsculas (A-Z) y del 37 al 61 las minúsculas (a-z)

```python
# función para hacer una clasificacón
# 0: números del 0 - 9    (0-9)
# 1: letras A - Z         (10-35)
# 2: letras a - z         (36-61)

def clasif(x):
    if 0<=x<=9: return 0
    elif 10<=x<=35: return 1
    elif 36<=x<=61: return 2

df_clasif = df.copy()
df_clasif['type'] = df_clasif['e'].apply(clasif)
df_clasif.head()
```

```python
numbers = df_clasif[df_clasif['type'] == 0]
lower = df_clasif[df_clasif['type'] == 1]
upper = df_clasif[df_clasif['type'] == 2]
```

* Now, try to train a classifier model to predict the uppercases. Use every single model you know for classification.

```python
from sklearn.linear_model import SGDClassifier as SGDC
from sklearn.neighbors import KNeighborsClassifier as KNNC
from sklearn.naive_bayes import MultinomialNB as MNB
#from sklearn.ensemble import GradientBoostingClassifier as GBC
from sklearn.svm import SVC
#from xgboost import XGBClassifier as XGBC
#from catboost import CatBoostClassifier as CATC
#from lightgbm import LGBMClassifier as LGBMC

models_lst = [SGDC, SVC, MNB, KNNC]
```

```python
def clasifier_tester(m_lst, data, target):
    X_train, X_test, y_train, y_test = tts(data.loc[:, data.columns != target], data[target])
    
    for i in tqdm(m_lst):
        model = i().fit(X_train, y_train)
        
        y_pred = model.predict(X_test)
        
        print('Modelo:',re.findall(r'\b(\w+)\b\'\>',str(i))[0]);
        print('Accuracy score:', accuracy_score(y_pred,y_test),'\n');
```

```python
#clasifier_tester(models_lst, upper, 'e')
```

```python
X_train, X_test, y_train, y_test = tts(upper.loc[:, upper.columns != 'e'], upper['e'], random_state = 1)
sgdc_model = SGDC().fit(X_train, y_train)
svc_model = SVC().fit(X_train, y_train)
mnb_model = MNB().fit(X_train, y_train)
knnc_model = KNNC().fit(X_train, y_train)
```

```python
sgdc_predict = sgdc_model.predict(X_test)
svc_predict = svc_model.predict(X_test)
mnb_predict = mnb_model.predict(X_test)
knnc_predict = knnc_model.predict(X_test)
```

```python
print('SGDC',accuracy_score(sgdc_predict,y_test))
print('SVC',accuracy_score(svc_predict,y_test))
print('MNB',accuracy_score(mnb_predict,y_test))
print('KNNC',accuracy_score(knnc_predict,y_test))
```

```python
# El mejor modelo para hacer esta clasificación es el SVC
svc_data = {'REAL':y_test, 'PREDICTION':svc_predict}
upper_df = pd.DataFrame(svc_data)
upper_relation = pd.crosstab(upper_df.REAL, upper_df.PREDICTION, rownames=['REAL'], colnames=['PREDICTION'])

sns.heatmap(upper_relation);
```

```python
sns.heatmap(confusion_matrix(y_test, svc_predict));
```
